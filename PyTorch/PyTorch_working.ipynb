```python
import torch
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace = True)
```

```python
df
```

```python
x_train , x_test , y_train , y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size =0.2)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_train)
```

```python
en = LabelEncoder()
y_train = en.fit_transform(y_train)
y_test = en.transform(y_test)
```

```python
x_train_tensor = torch.from_numpy(x_train)
x_test_tensor = torch.from_numpy(x_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)


```

```python
class SimpleNN :

  def __init__(self,x):
    self.weights = torch.rand(x_train.shape[1],1,dtype=torch.float64,requires_grad=True)
    self.bias = torch.rand(1,dtype=torch.float64,requires_grad=True)

  def forward(self,x):
    z = torch.matmul(x,self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss(self,y_pred,y):
    #to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred,epsilon,1-epsilon)
    loss = -(y*torch.log(y_pred)+(1-y)*torch.log(1-y_pred)).mean()
    return loss


```

```python
learning_rate = 0.1
epochs = 25

model = SimpleNN(x_train_tensor)

for epoch in range(epochs):

  y_pred = model.forward(x_train_tensor)
  loss = model.loss(y_pred,y_train_tensor)
  loss.backward()

  with torch.no_grad():
    model.weights-=learning_rate*model.weights.grad
    model.bias-=learning_rate*model.bias.grad

  model.weights.grad.zero_()
  model.bias.grad.zero_()

  print(f'epoch : {epoch + 1} , loss : {loss.item()}')

```
